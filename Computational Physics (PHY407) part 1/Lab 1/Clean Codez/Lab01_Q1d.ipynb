```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

QUESTION 1d

```python
#1. Define our constants
Ms = 1
G = 39.5
alpha = 0.01

#2. Pick a value for ∆t, as well as a time period to integrate over.
dt = 0.0001
t = 1
```

```python
#3. Set values for initial conditions: vx0, vy0, x0, y0, t0.
x = 0.47
y = 0.0
vx = 0.0
vy = 8.17
```

```python
#Define array for time values
time = np.zeros(int(t/dt))

#Define arrays for velocity and position values
vx_arr = np.zeros(int(t/dt))
vy_arr = np.zeros(int(t/dt))
x_arr = np.zeros(int(t/dt))
y_arr = np.zeros(int(t/dt))

#Store intitial conditions
vx_arr[0] = vx
vy_arr[0] = vy
x_arr[0] = x
y_arr[0] = y
time[0] = 0


for i in range(time.size-1):
    time[i] = time[i-1] + dt
    
    #5. Define r as √(x^2 + y^2).
    r = np.sqrt((x_arr[i]**2)+(y_arr[i]**2))
    
    #6. Using the Euler-Cromer method, update the velocity.
    vx_arr[i+1] = vx_arr[i] - ((G*Ms*x_arr[i])*(1+alpha/r**2)/(r**3))*dt
    vy_arr[i+1] = vy_arr[i] - ((G*Ms*y_arr[i])*(1+alpha/r**2)/(r**3))*dt
    
    #7. Use this updated velocity to update the position.
    x_arr[i+1] = x_arr[i] + vx_arr[i+1]*dt
    y_arr[i+1] = y_arr[i] + vy_arr[i+1]*dt
    
#8. Repeat steps 5 and 6 until the chosen time period is integrated over (This is completed by the for loop)

#9. Plot velocity vs. time and x vs. y.
plt.plot(time[0:-1], vy_arr[0:-1], label = "y velocity")
plt.plot(time[0:-1], vx_arr[0:-1], label = "x velocity")
plt.title("Velocity vs. Time For Mercury", fontsize = 16)
plt.xlabel("Time (years)", fontsize = 13)
plt.ylabel("Velocity (AU/Year)", fontsize = 13)
plt.legend(loc="lower right")
```

```python
plt.plot(x_arr, y_arr)
plt.title("X Position vs. Y Position for Mercury", fontsize = 16)
plt.xlabel("X (AU)", fontsize = 13)
plt.ylabel("Y (AU)", fontsize = 13)
```
