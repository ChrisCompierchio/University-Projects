```python
#import libraries
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define an array to store the N values
num = np.arange(400)

#Define arrays to stroe the times for each N value for each method of multiplication
loop_times = np.zeros(num.size)
dot_times = np.zeros(num.size)

#Loop from 2 to the size of the num array
for N in range(2, num.size): 
    
    #Create the NxN matrices
    A = np.ones([N, N], float)*3
    B = np.ones([N, N], float)*2
    C = np.ones([N,N], float)

    #start timer
    start_time = time.time()
    
    #Loop through matrices A and B to multiply them together and store them in matrix C
    for i in range(N):
        for j in range(N):
            for k in range(N):
                C[i,j] += A[i,k]*B[k,j]
    #end timer
    end_time = time.time()

    #store the time it took th loop to run
    loop_times[N] = end_time-start_time
    
    #start the timer
    start_time2 = time.time()
    #compute the matrix multiplication of A and B using np.dot()
    C = np.dot(A,B)
    #end the timer
    end_time2 = time.time()
    
    #store the times it took to multiply the matrices
    dot_times[N] = end_time2 - start_time2
```

```python
#Plot the times it took the loop to multiply the matrices vs N
plt.plot(num,loop_times)
plt.title("Time it takes Loop to Run N times", fontsize = 15)
plt.xlabel("N", fontsize = 13)
plt.ylabel("Time", fontsize = 13)
```

```python
#Plot the times it took the loop to multiply the matrices vs N^3
plt.plot(num**3, loop_times)
plt.title("Time it takes Loop to Run N^3 times", fontsize = 15)
plt.xlabel("N^3", fontsize = 13)
plt.ylabel("Time", fontsize = 13)
```

```python
#Plot the times it took the np.dot() function to multiply the matrices vs N
plt.plot(num, dot_times)
plt.title("Time it takes np.dot to Run N times", fontsize = 15)
plt.xlabel("N", fontsize = 13)
plt.ylabel("Time", fontsize = 13)
```

```python
#Plot the times it took the np.dot() function to multiply the matrices vs N^3
plt.plot(num**3, dot_times)
plt.title("Time it takes Loop to Run N^3 times", fontsize = 15)
plt.xlabel("N^3", fontsize = 13)
plt.ylabel("Time", fontsize = 13)
```
