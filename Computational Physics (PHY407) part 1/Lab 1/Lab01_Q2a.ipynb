```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

QUESTION 2a

```python
#1. Define our constants: G as 39.5AU^3 M^−1 S^yr−2, Ms as 1 solarmass, alpha = 0.01AU^2.
Ms = 1
G = 39.5
alpha = 0.01

Mj = (1*10**-3)*Ms
aj = 5.2

#2. Pick a value for ∆t, as well as a time period to integrate over.
dt = 0.0001
t = 10
```

```python
#3. Set values for initial conditions
xj = 5.2
yj = 0.0
vxj = 0.0
vyj = 2.63


xe = 1.0
ye = 0.0
vxe = 0.0
vye = 6.18
```

```python
#Define array for time values
time = np.zeros(int(t/dt))

#Define arrays for velocity and position values
vxj_arr = np.zeros(int(t/dt))
vyj_arr = np.zeros(int(t/dt))
xj_arr = np.zeros(int(t/dt))
yj_arr = np.zeros(int(t/dt))

vxe_arr = np.zeros(int(t/dt))
vye_arr = np.zeros(int(t/dt))
xe_arr = np.zeros(int(t/dt))
ye_arr = np.zeros(int(t/dt))

#Store intitial conditions
vxj_arr[0] = vxj
vyj_arr[0] = vyj
xj_arr[0] = xj
yj_arr[0] = yj

vxe_arr[0] = vxe
vye_arr[0] = vye
xe_arr[0] = xe
ye_arr[0] = ye

time[0] = 0

for i in range(time.size-1):
    time[i] = time[i-1] + dt
    
    #5. Define r as √(x^2 + y&2).
    rje = np.sqrt((xj_arr[i]-xe_arr[i])**2 + (yj_arr[i]-ye_arr[i])**2)
    rjs = np.sqrt((xj_arr[i]**2) + yj_arr[i]**2)
    res = np.sqrt((xe_arr[i]**2) + ye_arr[i]**2)
    
    #6. Using the Euler-Cromer method, update the velocity.
    vxj_arr[i+1] = vxj_arr[i] - ((G*Ms*xj_arr[i])/(rjs**3))*dt
    vyj_arr[i+1] = vyj_arr[i] - ((G*Ms*yj_arr[i])/(rjs**3))*dt
    
    #7. Use this updated velocity to update the position.
    xj_arr[i+1] = xj_arr[i] + vxj_arr[i+1]*dt
    yj_arr[i+1] = yj_arr[i] + vyj_arr[i+1]*dt
    
    #6. Using the Euler-Cromer method, update the velocity.
    vxe_arr[i+1] = vxe_arr[i] - ((G*Ms*xe_arr[i])/(res**3))*dt
    vye_arr[i+1] = vye_arr[i] - ((G*Ms*ye_arr[i])/(res**3))*dt
    
    #7. Use this updated velocity to update the position.
    xe_arr[i+1] = xe_arr[i] + vxe_arr[i+1]*dt
    ye_arr[i+1] = ye_arr[i] + vye_arr[i+1]*dt
    
#8. Repeat steps 5 and 6 until the chosen time period is integrated over. (for loop will do this)

#9. Plot velocity vs. time and x vs. y.
plt.plot(time[0:-1], vyj_arr[0:-1], label = "y velocity")
plt.plot(time[0:-1], vxj_arr[0:-1], label = "x velocity")
plt.title("Velocity vs Time For Mercury", fontsize = 16)
plt.xlabel("Time (years)", fontsize = 13)
plt.ylabel("Velocity (AU/Year)", fontsize = 13)
plt.legend(loc="lower right")
```

```python
plt.plot(xj_arr, yj_arr, label = "Jupiter Orbit")
plt.plot(xe_arr, ye_arr, label = "Earth Orbit")
plt.scatter(0,0, s = 15, color = "yellow", label = "Sun")

plt.title("X Position vs Y Position for Earth and Jupiter", fontsize = 16)
plt.xlabel("X (AU)", fontsize = 13)
plt.ylabel("Y (AU)", fontsize = 13)

plt.legend(loc="upper left")
```
