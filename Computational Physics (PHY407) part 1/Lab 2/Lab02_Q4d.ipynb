```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define u
u = np.linspace(0.98,1.02,500)
```

```python
#Calculate f
f = u**8/((u**4)*(u**4)) 
#Calculate and output the standard deviation for f
std = np.std(f)
std
```

```python
#Plot results
plt.plot(u, f-1)

plt.rc('font', size=15)
plt.title("f-1 Vs u")
plt.rc('font', size=15)
plt.xlabel("u")
plt.rc('font', size=15)
plt.ylabel("f-1")
```
