This program will compare a function with its expanded form.

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define u
u = np.linspace(0.98,0.984,500)
```

```python
#Calculate p and q
p = (1-u)**8
q = 1-8*u+(28*u**2)-(56*u**3)+(70*u**4)-(56*u**5)+(28*u**6)-(8*u**7)+u**8
```

```python
#Define C
C = 10**(-16)
#Store the size of the u array
N = u.size
#Calculate the mean of u
mean = np.mean(u**2)
```

```python
#calculate fractional error
frac_err = (C/np.sqrt(N))*(np.sqrt(mean)/np.mean(u))
frac_err
```

```python
#Plot results
plt.plot(u, abs(p-q)/abs(p))
plt.rc('font', size=15)
plt.title("Error Between p and q")
plt.rc('font', size=15)
plt.xlabel("u")
plt.rc('font', size=15)
plt.ylabel("Percent x100")
plt.axhline(y=1, color='red', label = '100%')
plt.legend(loc="best")
```
