This program will graphically compare a function with its expanded form.

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define u
u = np.linspace(0.98,1.02,500)
```

```python
#Calculate p and q
p = (1-u)**8
q = 1-8*u+(28*u**2)-(56*u**3)+(70*u**4)-(56*u**5)+(28*u**6)-(8*u**7)+u**8
```

```python
#Plot results
plt.plot(u, q, label = "q")
plt.plot(u, p, label = "p")

plt.title("q(u) and p(u)")
plt.xlabel("u")
plt.legend(loc="lower left")
```
