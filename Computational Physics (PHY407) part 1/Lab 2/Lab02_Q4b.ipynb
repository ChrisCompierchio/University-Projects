This program will compare a function with its expanded form.

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#define u
u = np.linspace(0.98,1.02,500)
```

```python
#Define p and q
p = (1-u)**8
q = 1-8*u+(28*u**2)-(56*u**3)+(70*u**4)-(56*u**5)+(28*u**6)-(8*u**7)+u**8
```

```python
#Plot results
plt.plot(u, p-q)

plt.title("p(u)-q(u)")
plt.xlabel("u")
```

```python
#Plot histogram of results
plt.hist(p-q)
plt.title("Histogram of p(u)-q(u)")
```

```python
#Calculate standard deviation using numpy.std
np.std(p-q)
```

```python
#Calculate standard deviation using the formula given in the handout
C = 10**(-16)
N = 500
mean = (np.sum(u)/N)**2

std = C*np.sqrt(N)*np.sqrt(mean)
std
```
