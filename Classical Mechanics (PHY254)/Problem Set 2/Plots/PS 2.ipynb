```python
#import libraries
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.arange(0.8, 3.0, 0.01)
V = ((1/(r**12))-(2/(r**6)))
F = -(12/r**7)+(12/r**13)
```

```python
plt.plot(r,V, label = "potential")
plt.plot(r,F, label = "Force")
plt.ylim(-10, 7)


plt.title("Chris Compierchio's Force and Potential vs Radial Separation Plot for Îµ, rm = 1")
plt.ylabel("Force(N) and Potential(J)")
plt.xlabel("Radial Seperation (m)")


##############
r1 = np.arange(0.85, 0.96, 0.05)


V1 = ((1/(r1**12))-(2/(r1**6)))
plt.scatter(r1, V1, color = "green", label = "x=0.85, 0.9, 0.95")

plt.legend(loc = "upper right")
```
