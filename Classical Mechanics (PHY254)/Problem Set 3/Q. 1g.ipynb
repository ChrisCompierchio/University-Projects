```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Q = 1

#set F0, m, and omega_0 to 1 for simplicity 
omegad = np.linspace(0,2,1000)
P = ((2*omegad**2)/Q)*((1/(((1-omegad**2)**2))-((2*omegad/Q)**2)))
```

```python
plt.plot(omegad,P)
plt.ylabel("Average Power (watts)")
plt.xlabel("Driving Frequency (1/s)")
plt.title("Average power vs Driving Frequency in a Damped Spring")
```
