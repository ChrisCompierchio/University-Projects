```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.signal import argrelextrema
from scipy import optimize
```

```python
fh1a, fh1b = np.loadtxt("franck hertz.txt", delimiter = "\t", skiprows = 2, unpack = True)
fh2a, fh2b = np.loadtxt("franck hertz 2.txt", delimiter = "\t", skiprows = 2, unpack = True)
fh3a, fh3b = np.loadtxt("franck hertz 3.txt", delimiter = "\t", skiprows = 2, unpack = True)
fh4a, fh4b = np.loadtxt("franck hertz 4.txt", delimiter = "\t", skiprows = 2, unpack = True)
fh5a, fh5b = np.loadtxt("franck hertz 5.txt", delimiter = "\t", skiprows = 2, unpack = True)
```

```python
plt.plot(fh1a, fh1b)
plt.title("Current vs Accelerating Voltage")
plt.xlabel("Accelerating Voltage (V)")
plt.ylabel("Current (A)")

minima1x = []
minima1y = []

for i in range(fh1b.size-1):
    if fh1a[i] > 5:
        if i == 0:
            i = 1
        if fh1b[i] < fh1b[i+1] and fh1b[i] < fh1b[i-1]:
            minima1x.append(fh1a[i])
            minima1y.append(fh1b[i])
plt.scatter(minima1x, minima1y, color = "orange")
plt.xticks(np.arange(-5,35, step = 2))
plt.axvline(17.1, color = "black")
plt.axvline(22.3, color = "black")
```
