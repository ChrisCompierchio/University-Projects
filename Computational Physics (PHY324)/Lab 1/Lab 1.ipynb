```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import *
import scipy as sp
import matplotlib.mlab as mlab
import scipy.integrate as integrate
from scipy.optimize import curve_fit
```

```python
therm_x, therm_y, therm_2 = np.loadtxt("Thermistor.csv", skiprows = 2, delimiter = ',', unpack = True)
fseven_x, fseven_y = np.loadtxt("47 Ohm.csv", skiprows = 1, delimiter = ',', unpack = True)
thou_x, thou_y = np.loadtxt("1000 Ohm.csv", skiprows = 1, delimiter = ',', unpack = True)
germ_x, germ_y = np.loadtxt("Germanium.csv", skiprows = 1, delimiter = ',', unpack = True)
sili_x, sili_y = np.loadtxt("Silicon.csv", skiprows = 1, delimiter = ',', unpack = True)
z_x, z_y = np.loadtxt("Zener.csv", skiprows = 2, delimiter = ',', unpack = True)
```

```python
plt.figure()
plt.plot(fseven_x, fseven_y, label = "47 Ohm Resistor")
plt.plot(thou_x, thou_y, label = "1000 Ohm Resistor")
plt.legend(loc = "best")
plt.xlabel("Voltage (mV)")
plt.ylabel("Current (mA)")
plt.title("I vs V for Multiple Two-Terminal Devices")
plt.errorbar(fseven_x, fseven_y, xerr = 500, yerr = 25, fmt = ".")
plt.errorbar(thou_x, thou_y, xerr = 500, yerr = 25, fmt = ".")

plt.figure()
plt.plot(z_x, z_y, label = "Zener")
plt.plot(sili_x, sili_y, label = "Silicon")
plt.plot(germ_x, germ_y, label = "Germanium")
plt.errorbar(z_x, z_y, xerr = 200, yerr = 25, fmt = ".")
plt.errorbar(sili_x, sili_y, xerr = 200, yerr = 25, fmt = ".")
plt.errorbar(germ_x, germ_y, xerr = 200, yerr = 25, fmt = ".")
plt.legend(loc = "best")
plt.xlabel("Voltage (mV)")
plt.ylabel("Current (mA)")
plt.title("I vs V for Multiple Two-Terminal Devices")
```
