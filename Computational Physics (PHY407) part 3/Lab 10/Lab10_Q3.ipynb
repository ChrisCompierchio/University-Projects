This program will evalute an integral using various monte carlo methods

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as c
plt.rcParams.update({'font.size': 16})
```

PART A

```python
#set variables
a = 0
b = 1
N = 10000
num = 100
#set array to store integral values
Is = np.zeros(num)

#define the function
def f(x):
    n = (x**(-0.5))/(1+np.e**x)
    return n

#Evaluate integral using Mean method
for i in range( num):
    I=0
    for j in range(1,N):
        I += f(np.random.random())/N
    I = (b-a) * I
    Is[i] = I
    print(I)
```

PART B

```python
#create array to store integrals
Js = np.zeros(num)

#define the w function
def w(x):
    return x**(-0.5)

#calculate the integrals using the importance method
for i in range( num):
    J=0
    for j in range(1,N):
        J += f(np.random.random())/(N*w(np.random.random()))
    J =  w(np.random.random())* J
    Js[i] = J
    print(J)
```

PART C

```python
#plot the histogram for the mean method
plt.hist(Is, 10, range=[.8,.88])
plt.title("Histogram for Mean Method")
plt.show()
```

```python
#plot the histogram for the importance method
plt.hist(Js, 10, range=[.8,.88])
plt.title("Histogram for Importance Method")
plt.show()
```
