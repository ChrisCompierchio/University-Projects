PART A

```python
"""
This program simulates Brownian motion in the presence of walls
Note that the physical behaviour would be to stick to walls,
which is the purpose of Q1a.
Author: Nico Grisouard, University of Toronto
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
def nextmove(x, y):
    """ randomly choose a direction
    0 = up, 1 = down, 2 = left, 3 = right"""
    direction =  np.random.randint(0,4)
    if direction == 0:  # move up
        y += 1
    elif direction == 1:  # move down
        y -= 1
    elif direction == 2:  # move right
        x += 1
    elif direction == 3:  # move left
        x -= 1
    else:
        print("error: direction isn't 0-3")
    return x, y
font = {'family': 'DejaVu Sans', 'size': 14}  # adjust fonts
rc('font', **font)
# %% main program starts here ------------------------------------------------|
# YOU NEED TO FINISH IT!
plt.ion()
Lp = 101  # size of domain
Nt = 5000  # number of time steps
# arrays to record the trajectory of the particle
centre_point = (Lp-1)//2  # middle point of domain
xp = np.zeros(Nt)
yp = np.zeros(Nt)
#store middle point
xp[0] = centre_point
yp[0] = centre_point

#make time array
times = np.zeros(Nt)

#calculate next moves
for i in range(1,Nt):
    xp[i], yp[i] = nextmove(xp[i-1], yp[i-1])
    times[i] = i
    # AND OFF YOU GO!

#plot random walk
plt.plot(xp, yp)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random Walk: X vs Y")
plt.show()

#plot x vs time
plt.plot(times, xp)
plt.xlabel("t")
plt.ylabel("X")
plt.title("Random Walk: X vs t")
plt.show()

#plot y vs time
plt.plot(times, yp)
plt.xlabel("t")
plt.ylabel("Y")
plt.title("Random Walk: Y vs t")
plt.show()
```

```python
# %% main program starts here ------------------------------------------------|
# YOU NEED TO FINISH IT!
plt.ion()
Lp = 101  # size of domain
Nt = 50  # number of time steps
num_particles = 5 #number of particles
# arrays to record the trajectory of the particle
centre_point = (Lp-1)//2  # middle point of domain
xp = np.zeros(Nt)
yp = np.zeros(Nt)
#store centre point
xp[0] = centre_point
yp[0] = centre_point

#make time array
times = np.zeros(Nt)
#loop through number of particles
for j in range(num_particles):
    centre_point = (Lp-1)//2  # middle point of domain
    
    #set initial positions of the particles
    if j == 0:
        xp[0] = centre_point
        yp[0] = centre_point
        
    elif j>0:
        xp[0] = xp[-1]
        yp[0] = yp[-1]
    #plot intial position
    plt.scatter(xp[0], yp[0], color = "green", label = "starting point")
    #set times
    times = np.zeros(Nt)
    
    #calculate next moves
    for i in range(1,Nt):
        xp[i], yp[i] = nextmove(xp[i-1], yp[i-1])
        #stop loop if particle hits the centre or a wall
        if xp[i] == 0 and yp[i]== 0 or xp[i] == Lp or yp[i] == Lp:
            break
        #store times
        times[i] = i
        # AND OFF YOU GO!
    #plot end position
    plt.scatter(xp[-1], yp[-1], color = "red", label = "end point")
    
    #Plot random walk
    plt.plot(xp, yp)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Random Walk: X vs Y For 5 Particles")
#plt.legend(loc="best")
plt.show()
```
