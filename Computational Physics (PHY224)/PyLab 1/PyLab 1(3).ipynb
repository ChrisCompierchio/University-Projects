```python
#import libraries
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#declare a timestep variable
dt = 0.01
#declare an initial time
t = 0
#declare a period (although my TA said this is not necessary so I did not use it)
T = 10
#store the calculated spring constant
k = 14.7
#store the recorded mass
m = 0.2003
#store the amplitude (in m)
A = 0.0968
#store the equilibrium point (also in m)
eq = .274
#calculate and store the andular frequency
omega = np.sqrt(k/m)
```

```python
#create an array to store 200 time intervals
ti = np.zeros(1000)

#create an index variable to help run through the loop
#index = 0
ti[0] = 0
#begin the time stepping loop
for m in range(len(ti)-1):
    #increase the time at the specified index by dt and store it.
    ti[m+1]=ti[m]+dt
```

```python
#create an array for position and velocity
y_vals = np.zeros(1000)
v_vals = np.zeros(1000)

#set the inital value for the position as the amplitude A
y_vals[0] = A

#begin the loop to calculate position and velocity at each time interval
for i in range(0,y_vals.size-1):
    y_vals[i+1] = y_vals[i] + dt*v_vals[i]
    v_vals[i+1] = v_vals[i] - (dt*(omega**2)*y_vals[i])
```

```python
#plot position vs time
plt.scatter(ti[0:1000], y_vals)
plt.xlabel("time(s)")
plt.ylabel("position(m)")
plt.title("position vs time of mass on a spring")
```

```python
#plot velocity vs Time
plt.scatter(ti[0:1000], v_vals)
plt.xlabel("Time(s)")
plt.ylabel("Velocity(m/s)")
plt.title("velocity vs time for a mass on a spring")
```

```python
#plot position vs velocity
plt.scatter(y_vals, v_vals)
plt.xlabel("position(m)")
plt.ylabel("Velocity(m/s)")
plt.title("Velocity Vs Position for a mass on a spring")
```

```python
#begin the loop to calculate position and velocity at each time interval
for i in range(0,y_vals.size-1):
    y_vals[i+1] = y_vals[i] + dt*v_vals[i]
    v_vals[i+1] = v_vals[i] - (dt*(k/m)*y_vals[i])

#create an array to store the energy values
E_vals = np.zeros(1000)

#calculate energy at each time interval
for i in range(0,E_vals.size):
    E_vals[i] = (.5*m*(v_vals[i])**2) + (.5*k*(y_vals[i])**2)
```

```python
#plot energy vs time
plt.scatter(ti[0:1000], E_vals)
plt.xlabel("Time(s)")
plt.ylabel("Energy(J)")
plt.title("Energy vs time for a mass on a spring")
```

```python
#create arrays for the symplectic method values
y_vals_symp = np.zeros(1000)
v_vals_symp = np.zeros(1000)
E_vals_symp = np.zeros(1000)

#set inital position to A
y_vals_symp[0] = A

#calculate y, v and E using symplectic equations
for i in range(0,v_vals_symp.size-1):
    y_vals_symp[i+1] = y_vals_symp[i] + (dt*v_vals_symp[i])
    v_vals_symp[i+1] = v_vals_symp[i] - (dt*(k/m)*y_vals_symp[i+1])
for i in range(0,E_vals_symp.size):
    E_vals_symp[i] = (.5*m*((v_vals_symp[i])**2)) + (.5*k*((y_vals_symp[i])**2))
```

```python
#plot symplectic energy vs time
plt.scatter(ti, E_vals_symp)
plt.xlabel("Time(s)")
plt.ylabel("Total Energy(J)")
plt.title("Energy vs Time of a Mass on a String")
```

```python
for i in range(0,v_vals_symp.size-1):
    y_vals_symp[i+1] = y_vals_symp[i] + (dt*v_vals_symp[i])
    v_vals_symp[i+1] = v_vals_symp[i] - (dt*(omega**2)*y_vals_symp[i+1])
for i in range(0,E_vals_symp.size):
    E_vals_symp[i] = (.5*m*((v_vals_symp[i])**2)) + (.5*k*((y_vals_symp[i])**2))
#plot symplectic velocity vs symplectic position
plt.scatter(y_vals_symp, v_vals_symp)
plt.xlabel("Position(m)")
plt.ylabel("Velocity(m/s)")
plt.title("velocity vs position of a Mass on a String")
```

```python
#store damping coefficient
gamma = .374
```

```python
#create arrays for the damping values
y_vals_damp = np.zeros(1000)
v_vals_damp = np.zeros(1000)
E_vals_damp = np.zeros(1000)

#set inital position to A
y_vals_damp[0] = A

#calculate y and v using damping
for i in range(0,v_vals_damp.size-1):
    y_vals_damp[i+1] = y_vals_damp[i] + (dt*v_vals_damp[i])
    v_vals_damp[i+1] = v_vals_damp[i] - (dt*(omega**2))*y_vals_damp[i+1]-(gamma*v_vals_damp[i]*dt)
```

```python
#plot damped position vs time
plt.scatter(ti,y_vals_damp)
plt.xlabel("time(s)")
plt.ylabel("position(m)")
plt.title("position vs time for damped mass-spring system")
```

```python
for i in range(0,v_vals_damp.size-1):
    y_vals_damp[i+1] = y_vals_damp[i] + (dt*v_vals_damp[i])
    v_vals_damp[i+1] = v_vals_damp[i] - (dt*(k/m))*y_vals_damp[i+1]-(gamma*v_vals_damp[i]*dt)
for i in range(0,E_vals_damp.size):
    E_vals_damp[i]=(.5*m*(v_vals_damp[i])**2) + (.5*k*(y_vals_damp[i])**2)

#plot damped energy vs time
plt.scatter(ti,E_vals_damp)
plt.xlabel("time(s)")
plt.ylabel("Energy(J))")
plt.title("Energy vs time for damped mass-spring system")
```
