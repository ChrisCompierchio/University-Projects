```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import *
import scipy as sp
import matplotlib.mlab as mlab
import scipy.integrate as integrate
from scipy.optimize import curve_fit
import matplotlib.dates as mdates
import pandas as pd
```

```python
df = pd.read_csv("dailyMinTemp.csv", parse_dates = ['Date'])
df['temp'] = df['temp'].astype(float)

plt.figure(figsize=(8,4))
plt.plot(df.Date, df.temp)
plt.title("Daily Minimum Temperatures in Melbourne From 1981 To 1991")
plt.xlabel("Year")
plt.ylabel("Temperature (C)")
```

```python
time=np.linspace(1981, 1982,len(df.Date))
fit = np.polyfit(time, df.temp, 1)
detrended_data = df.temp - np.polyval(fit, time)

plt.plot(detrended_data)
plt.title("Detrended Temperature Data")
plt.xlabel("Days From January 1, 1981")
plt.ylabel("Temperature(C)")
```

```python
fft_data = np.fft.fft(detrended_data)
freq = np.fft.fftfreq(len(detrended_data))

plt.plot(np.abs(freq), np.abs(fft_data))
plt.xlim(0,0.02)

#plt.axvline(x=freq[10], color = "orange")
#plt.axvline(x=freq[20], color = "orange")
plt.title("FFT of Temperature Data")
plt.xlabel("Frequency 1/day")
plt.ylabel("Amplitude")

print(freq[10], freq[20])
```
